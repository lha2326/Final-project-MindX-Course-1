```python
#Loại bỏ cột thừa và tạo thành một table mới
anime_new = anime.drop(columns=['Synonyms', 'Japanese', 'English', 'Synopsis', 'Status', 'Start_Aired', 'End_Aired','Licensors', 'Themes', 'Demographics'], axis=1)
```

```python
anime_new
```

```python
#Liệt kê những dòng có cột Favorites mang giá trị 0 và đếm
anime_new.loc[anime_new['Favorites']==0]
```

```python
sum(anime_new['Favorites']==0)
```

```python
#Loại bỏ những dòng có Favorites = 0
anime_new.drop(anime_new.index[anime_new['Favorites']==0], inplace=True) 
anime_new.loc[anime_new['Favorites']==0]
```

```python
#Loại bỏ những dòng có dữ liệu Null
anime_new.dropna(inplace=True)
```
